```python
def check_score(lst):
    symbol = {'#' : 5, 'O' : 3, 'X' : 1, '!' : -1, '!!' : -3, '!!!' : -5}
    expression = ''
    for i in lst:
        for j in i:
            expression = expression + '+' + '(' + str(symbol[j]) + ')'
    print('The Expression : ',expression )
    if eval(expression) < 0 :
        return 0
    else:
        return eval(expression)
```

```python
check_score([
['#', '!'],
['!!', 'X']
])
```

```python
check_score([
['!!!', 'O', '!'],
['X', '#', '!!!'],
['!!', 'X', 'O']
])
```

```python
def combinations(*args):
    comb = 1
    for i in args:
        comb = comb*i
    return comb
```

```python
combinations(2, 3)
```

```python
combinations(3, 7, 4)
```

```python
combinations(2, 3, 4, 5)
```

```python
def prime(n):
    if (n==1):
        return False
    elif (n==2):
        return True;
    else:
        for x in range(2,n):
            if(n % x==0):
                return False
        return True
```

```python

def char_position(letter):
    return (ord(letter) - 97) + 1

def to_boolean_list(s):
    lst = []
    for i in range(len(s)):
        if char_position(s[i]) % 2 == 0:
            lst.append(False)
        else:
            lst.append(True)
    return lst
```

```python
to_boolean_list('deep')
```

```python

to_boolean_list('loves')
```

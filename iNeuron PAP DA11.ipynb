```python
def isPrime(num1): 
    if num1 > 1:
        for i in range(2,num1):
            if (num1%i) == 0:
                #print(num1,'is not a prime')                
                break
        else:
            return True
    else:
        return False
    
def filter_primes(lst):
    new_lst = [i for i in lst if isPrime(i)]
    return new_lst
```

```python
filter_primes([7, 9, 3, 9, 10, 11, 27])
```

```python
def pop(lst):
    length = len(lst)//2
    fw = bw = length
    while(bw > 0):
        lst[bw] = length 
        lst[fw] = length
        bw -= 1
        fw += 1
        length -= 1
    return lst
```

```python
pop([0, 0, 0, 0, 4, 0, 0, 0, 0])
```

```python
pop([0, 0, 0, 3, 0, 0, 0])
```

```python
pop([0, 0, 2, 0, 0])
```

```python
def loves_me(n):
    lst =[]
    for i in range(1,n+1):
        if i % 2 == 0:
            lst.append("Loves me not")
        else:
            lst.append("Loves me")

    lst[-1] = lst[-1].upper()
    return ", ".join(lst)
```

```python
loves_me(3)
```

```python
loves_me(6)
```

```python
loves_me(1)
```

```python
def f_l(w):
    for j in w:
        if j.isalpha():
            x=j
            # print(x)
    return x


def sort_by_letter(l):
    if len(l)>0:
        m = l[0]
        for i in range(len(l)-1):
            for j in range(i+1,len(l)):
                if ord(f_l(l[i])) > ord(f_l(l[j])):
                    temp = l[i]
                    l[i]=l[j]
                    l[j]=temp
    return l
```

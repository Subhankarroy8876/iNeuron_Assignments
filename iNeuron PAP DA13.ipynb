```python
def remove_letters (l,w):
    n=[]
    for i in w:
        n.append(i)
    for item in n:
        if item in l:
            l.remove(item)
    return l
```

```python
remove_letters(["d", "b", "t", "e", "a", "i"], "edabit")
```

```python
remove_letters(["s", "t", "r", "i", "n", "g", "w"], "string")
```

```python
remove_letters(["b", "b", "l", "l", "g", "n", "o", "a", "w"], "balloon")
```

```python
def blocks(l):
    a=5
    b=7
    if(l==1):
        return a
    else:
        for i in range (1,l):
            a=a+b
            b=b+1
        return a
```

```python

blocks(5)
```

```python
blocks(1)
```

```python
blocks(2)
```

```python
def my_sub(a,b):
    n=[1 for i in range(a)]
    m=[1 for i in range(b)]
    for item in n:
        if item in m:
            m.remove(item)
    return sum(m)
```

```python
my_sub(5, 9)
```

```python
my_sub(10, 30)
```

```python
my_sub(0, 0)
```

```python
def add_bill(r):
    n=[]
    l =list(r.split(","))
    l=[i for i in l if i[0] == 'd']
    l=[i[1:] for i in l if i[0] == 'd']
    for i in l:
        if "k" in i:
            i = i[:-1]
            i= int(i)*1000
            n.append(i)
        else:
            n.append(int(i))
    return sum(n)
```

```python
add_bill("p30,d2k,p60,d200,p360")
```

```python
add_bill("d20,p40,p60,d50")
```

```python
add_bill("p30,d20,p60,d150,p360")
```

```python
def flip_list(l):
    y = []
    for i in l:
        if type(i)!= list:
            listA = [str(i)]
            for elem in listA:
               t = elem.split(', ')
               y.append((t))
        else:
            y.append(i[0])
    return y
```

```python
flip_list([1, 2, 3, 4])
```

```python
flip_list([[5], [6], [9]])
```

```python
flip_list([])
```

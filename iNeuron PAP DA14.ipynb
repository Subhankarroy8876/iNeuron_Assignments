```python
def show_the_love(l):
    m= min(l)
    x= l.index(m)
    tf = 0
    for i in range(len(l)):
        if l[i] != m:
            tf=tf+l[i]*0.25
            l[i] = l[i]-(l[i]*0.25)
    l[x]=l[x]+tf
    return(l)
```

```python
show_the_love([4, 1, 4])
```

```python
show_the_love([16, 10, 8])
```

```python
show_the_love([2, 100])
```

```python
def letters(l1,l2):
    l1=set(l1)
    l2=set(l2)
    l3=[]
    c=""
    for i in l1:
        if i in l2:
            c += i
    c= ''.join(sorted(c))
    a= ''.join(sorted(l1.difference(l2)))
    b= ''.join(sorted(l2.difference(l1)))
    l3.append(c)
    l3.append(a)
    l3.append(b)
    return(l3)
```

```python
letters("sharp", "soap")
```

```python
letters("board", "bored")
```

```python
letters("happiness", "envelope")
```

```python
letters("kerfuffle", "fluffy")
```

```python
letters("match", "ham")
```

```python
def pairs(l):
    r = l[::-1]
    a=[]
    n=[]
    i = 0
    if len(l) %2 == 0:
        length = int(len(l)/2)
    else:
        length = int(len(l)/2) +1
    while length:
        a.append(l[i])
        a.append(r[i])
        length = length -1
        i = i+1
    i=0
    while i<len(a):
        n.append(a[i:i+2])
        i+=2
    return n
```

```python
def add_str_nums(a,b):
        
    if not a:
        a="0"
    if not b:
        b="0"
    try:  
        return str(int(a)+int(b))
    except:
        return "-1"
```

```python
add_str_nums("4", "5")
```

```python
add_str_nums("abcdefg", "3")
```

```python
add_str_nums("1", "")
```

```python
add_str_nums("1874682736267235927359283579235789257", "32652983572985729")
```

```python
def unmix(s):
    s=''.join([ s[x:x+2][::-1] for x in range(0, len(s), 2) ])
    return s
```

```python
unmix("123456")
```

```python
unmix("hTsii  s aimex dpus rtni.g")
```

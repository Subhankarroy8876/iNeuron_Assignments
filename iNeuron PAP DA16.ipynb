```python
def valid_rondo(s):
    letter = ord("A")
    next_letter = ord("B")
    l= len(s)
    if l<2:
        return False
    elif s[0] and s[l-1] != "A":
        return False
    else:
        for i in range(0,l-1,2):

            if s[i]!= "A":
                return False
            else:
                for i in range(1,l-1,2):

                    if ord(s[i]) != next_letter:
                        return False
                    else:
                        next_letter += 1
                return True
```

```python

def sentence_searcher(txt,s):
    t = txt.split(".")
    s=s.lower()
    for i in t:
        x=i.lower()
        if s in x:

            return i
        return "" 
    
txt = "I have a cat. I have a mat. Things are going swell."
```

```python
def sum_round(n):
    n=str(n)
    n=n[::-1]
    z = ""
    for i in range(len(n)):
        x = int(n[i])*(10**i) 
        if x != 0:
            z += str(x)
            z += " "
    return z
```

```python
sum_round(101)
```

```python
sum_round(1234)
```

```python
sum_round(54210)
```

```python
def multiplication_table(n):   
    t = 1
    tab = []
    tmp = []
    while t <= n:
        tmp = []
        for i in range(1,n+1):
            tmp.append(i*t)
        tab.append(tmp)
        t = t + 1
    return tab
```

```python
multiplication_table(1)
```

```python
multiplication_table(3)
```

```python
multiplication_table(5)
```

```python
def last(s):
    s=s.split(" ")[-1]
    n1=""
    for i in s:
        if (i.isalpha()) == True:
            n1+=i
    return n1
def does_rhyme(s1,s2):
    s1=last(s1)
    s2=last(s2)
    for i in range(len(s1),0,-1):
        if s1 in s2:
            return True
        else:
            return False
```

```python
does_rhyme("Sam I am!", "Green eggs and ham.")
```

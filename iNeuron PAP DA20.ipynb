```python
def count_lone_ones(num):
    count=0
    lst=[int(i) for i in str(num)]
    if lst[0]==1 and lst[1]!=1:
        count+=1
    if lst[-2]!=1 and lst[-1]==1:
        count+=1
    for i in range(1,len(lst)-2):
        if lst[i]==1:
            if lst[i-1]!=1 and lst[i+1]!=1:
                count+=1
    return count
```

```python
count_lone_ones(101)
```

```python
count_lone_ones(1191)
```

```python
count_lone_ones(1111)
```

```python
count_lone_ones(462)
```

```python
def printGrid(rows,cols):
    
    rows,cols=cols,rows
    arr=[]
    x=1
    for i in range(rows):
        col = []
        for j in range(cols):
            col.append(x)
            x+=1
        arr.append(col)
    l2 =[[row[i] for row in arr] for i in range(len(arr[0]))]
    return l2
```

```python
printGrid(3, 6)
```

```python
printGrid(5, 3)
```

```python
printGrid(4, 1)
```

```python
def min_miss_pos(lst):
    pos=[i for i in range(1,11)]
    for i in lst:
        if i <0:
            lst.remove(i)
    lst=list(set(pos).difference(lst))
    lst.sort()
    return lst[0]
```

```python

min_miss_pos([-2, 6, 4, 5, 7, -1, 1, 3, 6, -2, 9, 10, 2, 2])
```

```python
def pizza_points(customers, minOrders, minPrice):
    
    new=[]
    for cust in customers:
        x=[i for i in customers[cust] if i>minPrice]
        if len(x)>=minOrders:
            new.append(cust)
    return new
```

```python
customers = {
  "Batman": [22, 30, 11, 17, 15, 52, 27, 12],
  "Spider-Man": [5, 17, 30, 33, 40, 22, 26, 10, 11, 45]
}
pizza_points(customers, 5, 20)
```

```python
pizza_points(customers, 3, 10)
```

```python
pizza_points(customers, 5, 100)
```

```python
def fib_fast(num):
    count = 0
    sum = 0
    f1 , f2 = 0,1
    if num <= 0:
        print("Not a positive number")
    elif num == 1:
        print("nth Fibonacci series upto ",num, ":")
        
        return num
    else:
        print("nth Fibonacci series upto ",num, ":")
        while count <= num:        
            count+=1
            if count > num:
                return f1
            fth = f1 + f2
            f1,f2 = f2, fth
```

```python
def convert_to_hex(str):
 
    # Initialize final String
    hexa = []
    for i in range(len(str)):      
        ch = str[i]      
        in1 = ord(ch)
        part = hex(in1).lstrip("0x")
        hexa.append(part)
    return " ".join(hexa)
```

```python
convert_to_hex('hello world')
```

```python
convert_to_hex('Big Boi')
```

```python
convert_to_hex('Marty Poppinson')
```

```python
def uncensor(string, vowels):
    new = ''
    counter = 0
    for i in range(len(string)):
        if string[i] == '*':            
            new += vowels[counter]
            counter += 1
        else:
            new += string[i]
    return new
```

```python
uncensor('Wh*r* d*d my v*w*ls g*?', 'eeioeo')
```

```python
uncensor('abcd', '')
```

```python
uncensor('*PP*RC*S*', 'UEAE')
```

```python
from dns import reversename
from dns import resolver

def get_domain(ipAdd):

    domain_address = reversename.from_address(ipAdd)
    domain_name = str(resolver.resolve(domain_address,"PTR")[0])
    return domain_name
```

```python

get_domain('8.8.8.8')
```

```python
get_domain('8.8.4.4')
```

```python
def factorial(n):
    fact = 1    
    for num in range(2, n + 1):
        fact = fact * num
    return(fact)

def fact_of_fact(num):
    factorial_of_fact = 1
    if num < 0:
        print("Can not calculate Factorial of a negative number")
    elif num == 0:
        print("Factorial of zero is 1")
    else:
        for i in range(1,num+1):
            factorial_of_fact = factorial_of_fact * factorial(i)
    return "Factorial of {} is {}".format(num, factorial_of_fact)
```

```python
fact_of_fact(4)
```

```python

fact_of_fact(5)
```
